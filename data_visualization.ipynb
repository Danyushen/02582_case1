```python
import pandas as pd
import missingno as msno

%matplotlib inline
import matplotlib.pyplot as plt

# seaborn can be used to "prettify" default matplotlib plots by importing and setting as default
import seaborn as sns
sns.set() # Set searborn as default
```

```python
data = pd.read_csv('case1Data.csv')

y = data['y']
X = data.drop(columns=['y'])

categorical_cols = [col for col in X.columns if 'C_' in col]
numerical_cols = [col for col in X.columns if col not in categorical_cols]
```

```python
# Visualizations for Introduction
import seaborn as sns

# Distribution of target variable y
plt.figure(figsize=(8, 5))
sns.histplot(y, kde=True)
plt.title('Distribution of y')
plt.xlabel('y')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Correlation heatmap of numerical features (subset due to high dimensionality)
numerical_subset = data[numerical_cols]#.iloc[:, :20]  # select subset of features for visualization
corr_matrix = numerical_subset.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

# Boxplot of categorical variables vs target
fig, axs = plt.subplots(1, len(categorical_cols), figsize=(20, 5))
for i, cat in enumerate(categorical_cols):
    sns.boxplot(x=data[cat], y=y, ax=axs[i])
    axs[i].set_title(f'{cat} vs y')
plt.tight_layout()
plt.show()
```
