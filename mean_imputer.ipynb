```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.linear_model import ElasticNet, Ridge
```

```python
file_path = "case1Data.csv"
df = pd.read_csv(file_path)
```

```python
X = df.drop(columns=["y"])
y = df["y"]
```

```python
imputer = SimpleImputer(strategy="mean")
X_imputed = imputer.fit_transform(X)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
param_grid = {
    "alpha": np.logspace(-3, 3, 10),
    "l1_ratio": np.linspace(0.1, 0.9, 5)
}
```

```python
enet = ElasticNet(max_iter=5000)
grid_search = GridSearchCV(enet, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_scaled, y)
```

```python
best_enet = grid_search.best_estimator_
best_alpha = grid_search.best_params_["alpha"]
best_l1_ratio = grid_search.best_params_["l1_ratio"]
```

```python
rmse_scores = np.sqrt(-cross_val_score(best_enet, X_scaled, y, cv=5, scoring='neg_mean_squared_error'))
```

```python
print(f"Best Alpha: {best_alpha}")
print(f"Best L1 Ratio: {best_l1_ratio}")
print(f"Mean RMSE: {rmse_scores.mean():.2f}")
print(f"Standard Deviation of RMSE: {rmse_scores.std():.2f}")
```

# PCA and ElasticNet

```python
pca_data = pd.read_csv('pca_data.csv')
X_pca = pca_data.drop('y', axis=1)

# take the first 40 principal components
X_40 = X_pca.iloc[:, :40]

[n, p] = np.shape(X_40)
```

```python
# Include offset / intercept
off = np.ones(n) 
M = np.c_[off, X_40]
```

```python
param_grid = {
    "alpha": np.logspace(-3, 3, 10),
    "l1_ratio": np.linspace(0.1, 0.9, 5)
}
```

```python
enet = ElasticNet(max_iter=5000)
grid_search = GridSearchCV(enet, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(M, y)
```

```python
best_enet = grid_search.best_estimator_
best_alpha = grid_search.best_params_["alpha"]
best_l1_ratio = grid_search.best_params_["l1_ratio"]
```

```python
rmse_scores = np.sqrt(-cross_val_score(best_enet, M, y, cv=5, scoring='neg_mean_squared_error'))
```

```python
print(f"Best Alpha: {best_alpha}")
print(f"Best L1 Ratio: {best_l1_ratio}")
print(f"Mean RMSE: {rmse_scores.mean():.2f}")
print(f"Standard Deviation of RMSE: {rmse_scores.std():.2f}")
```

# PCA and Ridge

```python
param_grid = {"alpha": np.logspace(-3, 3, 10)}
ridge = Ridge()
grid_search = GridSearchCV(ridge, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(M, y)
```

```python
best_ridge = grid_search.best_estimator_
best_alpha = grid_search.best_params_["alpha"]
```

```python
rmse_scores = np.sqrt(-cross_val_score(best_ridge, M, y, cv=5, scoring='neg_mean_squared_error'))
```

```python
print(f"Best Alpha: {best_alpha}")
print(f"Mean RMSE: {rmse_scores.mean():.2f}")
print(f"Standard Deviation of RMSE: {rmse_scores.std():.2f}")
```
